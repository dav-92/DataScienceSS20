```python
#import an read big-mart data
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv(path+"/DATA/company_sales_data.csv") 
```

```python
#access data rows by name
plt.plot(data.toothpaste)
```

## Exercise 1
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: 

* X label name = Month Number
* Y label name = Total profit


```python
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])

axes.plot(data.month_number, data.total_profit, 'r')

axes.set_xlabel('Month Number')
axes.set_ylabel('Total profit')
axes.set_title('Total profit per month')
```

## Exercise 2
Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3


```python
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])

axes.plot(data.month_number, data.total_profit, 'r:', linewidth=3.00, marker='o')

axes.set_xlabel('Month Number')
axes.set_ylabel('Sold units number')
axes.set_title('Total profit per month')
axes.legend(["profit"], loc=4)
```

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).

```python
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])

axes.plot(data.month_number, data.facecream, linewidth=3.00, marker='o', label="Facecream")
axes.plot(data.month_number, data.facewash, linewidth=3.00, marker='o', label="Facewash")
axes.plot(data.month_number, data.toothpaste, linewidth=3.00, marker='o', label="Toothpaste")
axes.plot(data.month_number, data.bathingsoap, linewidth=3.00, marker='o', label="Bathingsoap")
axes.plot(data.month_number, data.shampoo, linewidth=3.00, marker='o', label="Shampoo")
axes.plot(data.month_number, data.moisturizer, linewidth=3.00, marker='o', label="Moisturizer")

axes.set_xlabel('Month Number')
axes.set_ylabel('Sold units number')
axes.set_title('Total sales')
axes.legend(loc=4)
```

## Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot.

```python
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])
axes.scatter(data.month_number, data.toothpaste)
axes.grid()
axes.set_xlabel('Month Number')
axes.set_ylabel('Sold units number')
axes.set_title('Toothpaste sales')
```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])
axes.bar(data.month_number-0.15, data.facecream, width=0.5, label="Facecream")
axes.bar(data.month_number+0.15, data.facewash, width=0.5, label="Facewash")
axes.set_xlabel('Month Number')
axes.xaxis.set_ticks(data.month_number)
axes.set_ylabel('Sold units number')
axes.set_title('Facecream and Facewash sales')
axes.legend(loc=4)
```

## Exercise 6
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

***Note:*** use ``data[item].sum()`` to compute sums over columns 

```python
fig = plt.figure()
axes = fig.add_axes([1,1,1,1])
axes.pie([data.facecream.sum(), data.facewash.sum(), data.toothpaste.sum(), data.bathingsoap.sum(), data.shampoo.sum(), data.moisturizer.sum()],
         labels=["Facecream", "Facewash", "Toothpaste", "Bathingsoap", "Shampoo","Moisturizer"],
         autopct='%1.1f%%')
axes.set_title('Sales')
axes.legend(loc=4)
```
